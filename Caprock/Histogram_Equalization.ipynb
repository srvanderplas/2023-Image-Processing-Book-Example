```python
import cv2
import matplotlib.pyplot as plt
import skimage
import os
from skimage.color import rgb2gray
from skimage import exposure
```

```python
src = cv2.imread(os.path.join("Caprock", "Caprock_color.png"))
```

```python
src = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
plt.imshow(src, "gray")
```

```python
dst = cv2.equalizeHist(src)
plt.imshow(dst, "gray")
```

```python
plt.hist(dst.ravel())
plt.hist(dst.ravel(), histtype='step', color='black', density = False)
plt.show()
```

```python
src = src - 50
plt.imshow(src, "gray")
```

```python
dst = cv2.equalizeHist(src)
plt.hist(dst.ravel())
plt.hist(dst.ravel(), histtype='step', color='black', density = False)
plt.show()
```

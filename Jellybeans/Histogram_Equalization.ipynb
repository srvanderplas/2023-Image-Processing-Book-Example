```python
import cv2
import matplotlib.pyplot as plt
import skimage
import os
from skimage.color import rgb2gray
from skimage import exposure
```

```python
datadir = os.path.join("../", "jellybean-data", "berry_blue")
all_ims = os.listdir(datadir)
```

```python
all_ims
```

```python
_, axs = plt.subplots(3, 2, figsize = (10,10))
axs = axs.ravel()
counter = 0
for file in all_ims[:3]:
    im = plt.imread(os.path.join(datadir, file))
    im = rgb2gray(im)
    axs[counter].imshow(im, "gray")
    counter += 1
    if counter == 5:
        axs[counter-1].set_xlabel("Unequalized Images", fontsize = 12, fontweight='bold')
#     plt.show()
#     plt.hist(im.ravel())
#     plt.show()
    img_eq = exposure.equalize_hist(im)
    axs[counter].imshow(img_eq, "gray")
#     counter += 1
#     plt.show()
#     axs[counter].hist(img_eq.ravel())
#     plt.show()
    counter += 1
    if counter == 6:
        axs[counter-1].set_xlabel("Equalized Images", fontsize = 12, fontweight='bold')
```

```python
samp_image = plt.imread(os.path.join(datadir, all_ims[0]))
```

```python
plt.imshow(samp_image)
plt.show()
```

```python
from skimage.color import rgb2gray
```

```python
samp_image_gray = rgb2gray(samp_image)
```

```python
plt.imshow(samp_image_gray, "gray")
plt.show()
```

```python
samp_image_gray_pixels =samp_image_gray.ravel()
```

```python
plt.hist(samp_image_gray_pixels)
plt.hist(samp_image_gray_pixels, histtype='step', color='black', density = False)
plt.show()
```

```python
plt.hist(samp_image_gray_pixels,cumulative=True, density=True)
plt.show()
```

```python
from skimage import exposure
```

```python
img_eq = exposure.equalize_hist(samp_image_gray)
```

```python
plt.imshow(img_eq, "gray")
plt.show()
```

```python
plt.hist(img_eq.ravel())
plt.hist(img_eq.ravel(), histtype='step', color='black', density = False)
plt.show()
```

```python
plt.hist(img_eq.ravel(),cumulative=True, density=True, bins=256)
plt.show()
```

```python
from skimage.filters import threshold_otsu
```

```python
# get the otsu thresholding
img_threshold = threshold_otsu(img_eq)
```

```python
# threshold the image
binary = img_eq > img_threshold
```

```python
plt.imshow(binary, "gray")
plt.show()
```

```python
plt.imshow(1-binary, "gray")
plt.show()
```
